### Landing errors
The following file contains the position of the landing platform and the position of the drone when landed in the simulated environment (centroid coordinates (x,y) and angle in radians). The data provided correspond to the tests of the three controllers (P, PI and PID). There are six columns per controller, three of them for the plaftorm's position an the other three for the drone's position. 


```python
file = "./data/landing_2.csv"
data = pd.read_csv(file)
```

```python
data
```

Converting the angle from radians to degrees.

```python
angle_columns = [i for i in data.columns if 'yaw' in i]
for column in angle_columns:
    data[column] = data[column] * 180 / math.pi
```

```python
data
```

A table with the absulote error and the standard deviation will be made using the position of the platform and the position of the drone for each controller. Remember that the angle was converted from radians to degrees in the cells above.

```python
land_df = pd.DataFrame(columns=["X_error_avg", "X_error_std",
                                "Y_error_avg", "Y_error_std",
                                "Angle_error_avg", "Angle_error_std"],
                       index=["P", "PI", "PID"])

land_df["X_error_avg"]["P"] = abs(data["P_platform_x"]-data["P_pos_x"]).mean()
land_df["X_error_avg"]["PI"] = abs(data["PI_platform_x"]-data["PI_pos_x"]).mean()
land_df["X_error_avg"]["PID"] = abs(data["PID_platform_x"]-data["PID_pos_x"]).mean()

land_df["Y_error_avg"]["P"] = abs(data["P_platform_y"]-data["P_pos_y"]).mean()
land_df["Y_error_avg"]["PI"] = abs(data["PI_platform_y"]-data["PI_pos_y"]).mean()
land_df["Y_error_avg"]["PID"] = abs(data["PID_platform_y"]-data["PID_pos_y"]).mean()

land_df["Angle_error_avg"]["P"] = abs(data["P_platform_yaw"]-data["P_pos_yaw"]).mean()
land_df["Angle_error_avg"]["PI"] = abs(data["PI_platform_yaw"]-data["PI_pos_yaw"]).mean()
land_df["Angle_error_avg"]["PID"] = abs(data["PID_platform_yaw"]-data["PID_pos_yaw"]).mean()

land_df["X_error_std"]["P"] = abs(data["P_platform_x"]-data["P_pos_x"]).std()
land_df["X_error_std"]["PI"] = abs(data["PI_platform_x"]-data["PI_pos_x"]).std()
land_df["X_error_std"]["PID"] = abs(data["PID_platform_x"]-data["PID_pos_x"]).std()

land_df["Y_error_std"]["P"] = abs(data["P_platform_y"]-data["P_pos_y"]).std()
land_df["Y_error_std"]["PI"] = abs(data["PI_platform_y"]-data["PI_pos_y"]).std()
land_df["Y_error_std"]["PID"] = abs(data["PID_platform_y"]-data["PID_pos_y"]).std()

land_df["Angle_error_std"]["P"] = abs(data["P_platform_yaw"]-data["P_pos_yaw"]).std()
land_df["Angle_error_std"]["PI"] = abs(data["PI_platform_yaw"]-data["PI_pos_yaw"]).std()
land_df["Angle_error_std"]["PID"] = abs(data["PID_platform_yaw"]-data["PID_pos_yaw"]).std()

land_df
```
